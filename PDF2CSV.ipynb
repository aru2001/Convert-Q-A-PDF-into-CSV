```python
from pdfminer.high_level import extract_text
import pandas as pd 
import re
```

```python
text = extract_text('Input.pdf')
```

```python
text
```

```python
text_split = text.split('\n')
for i in range(len(text_split)):
    print(text_split[i])
```

```python
text_split = text.split('Answer:')
text_split
```

```python
# import time
# start_time = time.time()

#text_split = text.split('Answer:')
#regular_exp = re.compile(r'[\n|\x0c][\s][\n|\x0c][0-9]+[\.|\,].*(((?:[\s]|\n).*){0,3})\?')

regular_exp = re.compile(r'[\n|\x0c][0-9]+[\.|\,].*(?:[\s]|\n).*(?:[\s]|\n).*\?')

question = []
answer = []

question.append(text_split[0])

for i in text_split:   
    
    matches = regular_exp.finditer(i)
    
    for match in matches:
        
        question.append(i[match.start():])
        answer.append(i[:match.start()])
        
answer.append(text_split[-1])

dict = {'question': question, 'answer': answer}  
     
df = pd.DataFrame(dict) 
  
df.to_csv('output.csv',index=False)

# print("--- %s seconds ---" % (time.time() - start_time))
```
